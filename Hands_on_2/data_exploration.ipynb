### ðŸ‘€ Data exploration
Here you will find a quick overview over the boston housing dataset that is integrated in keras.

```python
# Import keras import boston housing dataset
from keras.datasets import boston_housing

# Import pandas to easily show the data as table
import pandas as pd

# Import numpy to perform array tasks
import numpy as np

# Load the dataset from keras.datasets with load_data() function. The data is natively splitted into train and test.
(x_train, y_train), (x_test, y_test) = boston_housing.load_data()

##############################################
# Translation of the column names (our features)
##############################################
# CRIM - per capita crime rate by town
# ZN - proportion of residential land zoned for lots over 25,000 sq.ft.
# INDUS - proportion of non-retail business acres per town.
# CHAS - Charles River dummy variable (1 if tract bounds river; 0 otherwise)
# NOX - nitric oxides concentration (parts per 10 million)
# RM - average number of rooms per dwelling
# AGE - proportion of owner-occupied units built prior to 1940
# DIS - weighted distances to five Boston employment centres
# RAD - index of accessibility to radial highways
# TAX - full-value property-tax rate per $10,000
# PTRATIO - pupil-teacher ratio by town
# B - 1000(Bk - 0.63)^2 where Bk is the proportion of blacks by town
# LSTAT - % lower status of the population
############################################

# Specify some column names and print the head of the dataset
column_names = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT']

# Define a dataframe with pandas
df = pd.DataFrame(x_train, columns=column_names)

# Print the head of the dataframe
print(df.head())
```

### Get the Dimensions of the dataset

```python
# Print the shape of the train and test data to get the dimensions.
print(np.shape(x_train))
print(np.shape(x_test))
print(np.shape(y_train))
print(np.shape(y_test))


# In this case there are 404 Train Datapoints with 13 features each and 102 Test datapoints with 13 features each.
# The shape of the labels is equal to the length of x_train or x_test! 
```
